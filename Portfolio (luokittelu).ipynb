```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

data=pd.read_excel('http://taanila.fi/wine.xlsx')
```

```python
data.head()
```

```python
sns.pairplot(data, hue='Y')
```

```python
X = data.drop('Y', axis=1)
y = data['Y']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=5)
```

```python
from sklearn.neighbors import KNeighborsClassifier

#Mallin sovitus:
model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, y_train)

#Mallin mukaisten ennusteiden laskeminen opetusdatalle ja testidatalle:
y_train_model = model.predict(X_train)
y_test_model = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_train, y_train_model)
```

```python
accuracy_score(y_test, y_test_model)
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_train, y_train_model))
```

```python
print(confusion_matrix(y_test, y_test_model))
```
