```python
import pandas as pd

raw_data = {'Opiskelija' : ['5-oppilas', '4-oppilas', '3-oppilas', '2-oppilas', '1-oppilas', 'Minä'],
           'Aikasarjat_pisteet' : [10, 8, 6, 4, 1, 9],
           'Kuvailu_pisteet' : [9, 8, 4, 4, 2, 10], 
           'Mallit_pisteet' : [9, 7, 5, 3, 1, 8]}

df = pd.DataFrame(raw_data,columns = ['Opiskelija', 'Aikasarjat_pisteet', 'Kuvailu_pisteet', 'Mallit_pisteet'])

df
```

```python
df['Pisteet yhteensä'] = (df['Aikasarjat_pisteet'] + df['Kuvailu_pisteet'] + df['Mallit_pisteet'])
df
```

```python
df['Numero'] = (df['Pisteet yhteensä'] / 6)
df
```

```python
df.round(2)
```

```python
df.describe().round(1)
```

```python
df.corr()
```

```python
import seaborn as sns
sns.heatmap(df.corr(), annot = True)
```

```python
df['%-osuus kokonaispisteistä'] = (df['Pisteet yhteensä'] / 30 *100)
df.round(1)
```
